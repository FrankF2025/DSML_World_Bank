```python
# Time Series HDI with Slider (Adjusted for readability)
import dash
from dash import dcc, html, Output, Input
import plotly.express as px

# Extract years dynamically
year_columns = [col for col in data1.columns if col.startswith('hdi_')]
years = sorted([int(col.split('_')[1]) for col in year_columns if col.split('_')[1].isdigit()])

app = dash.Dash(__name__)

app.layout = html.Div([
    dcc.Slider(
        id='year-slider',
        min=min(years),
        max=max(years),
        value=min(years),
        marks={str(year): str(year) for year in years},
        step=1
    ),
    html.Div(id='selected-year', style={'fontSize': 20, 'marginTop': 10}),
    dcc.Graph(id='hdi-bar-chart')
])

@app.callback(
    Output('hdi-bar-chart', 'figure'),
    Output('selected-year', 'children'),
    Input('year-slider', 'value')
)
def update_chart(selected_year):
    year_column = f'hdi_{selected_year}'
    
    # Filter and prepare data
    df_filtered = data1[['country', year_column]].dropna()
    df_filtered = df_filtered.sort_values(by=year_column, ascending=True)  # Updated (Greatest to Least)
    
    # Create bar chart
    fig = px.bar(
        df_filtered,
        x=year_column,
        y='country',
        orientation='h',
        title=f'Human Development Index (HDI) - {selected_year}',
        labels={year_column: 'HDI', 'country': 'Country'},
    )

    # Adjust layout for readability
    fig.update_layout(
        xaxis=dict(range=[0, 1], tickmode='array', tickvals=[i/10 for i in range(11)]),
        height=2000,
        yaxis=dict(
            tickmode='array',
            tickvals=df_filtered['country'],
            ticktext=df_filtered['country'],
            tickfont=dict(size=10)  # Slightly larger for readability
        ),
        margin=dict(l=200)  # More space for country labels
    )
    
    return fig, f"Selected Year: {selected_year}"

if __name__ == '__main__':
    app.run_server(debug=True, port=8060) #Use different ports if you start getting errors.
```

```python
inv_columns = [col for col in merged_data.columns if col.endswith('_x')]
hdi_columns = [col for col in merged_data.columns if col.endswith('_y')]
years = sorted([int(col.split('_')[1]) for col in year_columns if col.split('_')[1].isdigit()])

# Define features (X) and target variable (y)
X = merged_data[inv_columns]  # WB_Investment amount and Year
y = merged_data[hdi_columns]  # Target: HDI values

# Split data into training (70%) and testing (30%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1010)
```

```python
# Train the regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Predict on test data
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R² Score: {r2}")
```
